### Step 2: Fix Wind variables (time-aware)
- Data Cleaning - Handling Missing Values Of 'Wind Speed (m/s)' Column Using Interpolation 
- Interpolation means estimating missing values that fall within the range of known values.
- Then we will use time-based interpolation since our data is time-series data.

```python
# Sort the data by City Name and Record Created At
Weather_Data = Weather_Data.sort_values(['City','UpdatedAt','Date'])
Weather_Data
```

### A) Now Handle Missing Values Of "Wind Speed"

```python
# Find rows where 'Wind Speed (m/s)' is still missing
Weather_Data[Weather_Data['WindSpeed']
            .isnull()][['WeatherCondition', 'WeatherDescription', 'WindSpeed', 'City', 'Date', 'UpdatedAt']]
```

```python
# See the 'Wind Speed (m/s)' column along with other relevant columns
Weather_Data[Weather_Data['WindSpeed']
            .notnull()][['WeatherCondition', 'WeatherDescription', 'WindSpeed', 'City', 'Date', 'UpdatedAt']]
```

```python
# Temporarily set datetime as index for time-based interpolation
Weather_Data = Weather_Data.set_index('UpdatedAt')
```

```python
# Do Time-based interpolation with filling remaining NaNs with median
# Interpolation meaning filling the missing values in a time series data by estimating them based on the values before and after the missing 
# data points. In this case, we are using time-based interpolation to fill in the missing 'Wind Speed (m/s)' values for each city. 
# If there are still any missing values after interpolation, we will fill those with the median value of 'Wind Speed (m/s)' for that city.
Weather_Data['WindSpeed'] = (
    Weather_Data.groupby('City')['WindSpeed']
      .transform(lambda x: x.interpolate(method='time')
                            .fillna(x.median()))
)
```

```python
# Restore DataFrame from index to get back its original structure
Weather_Data = Weather_Data.reset_index()
```

```python
# Check missing values
print(Weather_Data.isnull().sum())
```

```python
Weather_Data['WindSpeed'] = Weather_Data['WindSpeed'].round(2)
Weather_Data['WindSpeed']
```

### B) Now Handle Missing Values Of 'Wind Direction (deg)' Column Using circular interpolation

```python
# Find rows where 'Wind Direction (degrees)' is still missing
Weather_Data[Weather_Data['WindDirection']
            .isnull()][['WeatherCondition', 'WeatherDescription', 'WindDirection', 'City', 'Date', 'UpdatedAt']]
```

```python
# See the 'Wind Direction (degrees)' column along with other relevant columns
Weather_Data[Weather_Data['WindDirection']
            .notnull()][['WeatherCondition', 'WeatherDescription', 'WindDirection', 'City', 'Date', 'UpdatedAt']]
```

```python
# Now Handle Missing Values Of 'Wind Direction (deg)' Column Using circular interpolation
# Since wind direction is a circular variable (0 degrees is the same as 360 degrees),
# we will use circular interpolation to fill the missing values.

def circular_mean(series):
    vals = series.dropna()
    if len(vals) == 0:
        return series.fillna(0)
    rad = np.deg2rad(vals)
    sin_sum = np.sin(rad).mean()
    cos_sum = np.cos(rad).mean()
    
    # Calculate mean angle
    # arctan2 returns angle in radians, convert back to degrees
    # For example, if arctan2 returns -90 degrees, we convert it to 270 degrees
    mean_angle = np.rad2deg(np.arctan2(sin_sum, cos_sum)) % 360
    return series.fillna(mean_angle)

Weather_Data['WindDirection'] = (
    Weather_Data.groupby(['City','Date'])['WindDirection']
      .transform(circular_mean)
)
```

```python
# Step 1: City + Weather Condition median
Weather_Data['CloudCoverage'] = Weather_Data['CloudCoverage'].fillna(
    Weather_Data.groupby(['City','WeatherCondition'])['CloudCoverage']
                .transform('median')
)

# Step 2: City + Date median
Weather_Data['CloudCoverage'] = Weather_Data['CloudCoverage'].fillna(
    Weather_Data.groupby(['City','Date'])['CloudCoverage']
                .transform('median')
)

# Step 3: City median
Weather_Data['CloudCoverage'] = Weather_Data['CloudCoverage'].fillna(
    Weather_Data.groupby('City')['CloudCoverage']
                .transform('median')
)

# Step 4: Global median (final fallback)
Weather_Data['CloudCoverage'] = Weather_Data['CloudCoverage'].fillna(
    Weather_Data['CloudCoverage'].median()
)
```

```python
# Map AQI values to categories using range-based mapping
def aqi_category(aqi):
    if pd.isna(aqi):
        return None
    elif aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Satisfactory'
    elif aqi <= 200:
        return 'Moderate'
    elif aqi <= 300:
        return 'Poor'
    elif aqi <= 400:
        return 'Very Poor'
    else:
        return 'Severe'

Weather_Data['AQI_Category'] = Weather_Data['AQI'].map(aqi_category)
```
